```python
# PHASE 1: Upload Excel file and load data
from google.colab import files
import pandas as pd

uploaded = files.upload()  # Upload your Excel file manually

# Automatically get uploaded file name
import io
df = pd.read_excel(io.BytesIO(list(uploaded.values())[0]))

# Optional: Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()
df.head()
```

```python
# PHASE 2: Total number of students
num_students = len(df)
print(f"Total number of students: {num_students}")
```

```python
# Select subject columns
subjects = ["SQL", "JAVA", "AI", "PYTHON"]

# Calculate Total and Average
df["Total"] = df[subjects].sum(axis=1)
df["Average"] = df["Total"] / len(subjects)

# Attendance Status Rule
def attendance_fine(att):
    if att >= 80:
        return "Eligible"
    else:
        return "Not Eligible"

df["Attendance_Status"] = df["Attendance (%)"].apply(attendance_fine)

# Grade Calculation Rule
def calculate_grade(avg, att):
    if att < 75:
        return "UA"  # Unauthorized Absence
    elif avg >= 85:
        return "A+"
    elif avg >= 70:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 45:
        return "C"
    else:
        return "RA"

df["Grade"] = df.apply(lambda row: calculate_grade(row["Average"], row["Attendance (%)"]), axis=1)

# Pass/Fail Result based on attendance and marks
def get_result(avg, att):
    if att < 75 or avg < 45:
        return "FAIL"
    else:
        return "PASS"

df["Result"] = df.apply(lambda row: get_result(row["Average"], row["Attendance (%)"]), axis=1)

# 🎯 Rank only for students who passed and attendance ≥ 75
rank_eligible = df[(df["Attendance (%)"] >= 75) & (df["Average"] >= 45)]
df["Rank"] = None  # default None for all

df.loc[rank_eligible.index, "Rank"] = rank_eligible["Average"].rank(method="min", ascending=False).astype("Int64")

# Display final DataFrame
df
def highlight_result(val):
    if val == "PASS":
        return "color: light green; font-weight: bold"
    elif val == "FAIL":
        return "color: black; font-weight: bold"
    return ""

def highlight_grade(val):
    if val == "UA":
        return "color: black; font-weight: bold"
    elif val == "RA":
        return "color: red; font-weight: bold"
    else:
        return "color: green; font-weight: bold"

def highlight_name(val):
    return "color: sky blue; font-weight: bold"

# Apply all styles
df.style.applymap(highlight_name, subset=["Name"])\
        .applymap(highlight_result, subset=["Result"])\
        .applymap(highlight_grade, subset=["Grade"])

```

```python
import matplotlib.pyplot as plt

# Filter: Attendance ≥75 and Pass
eligible_df = df[(df["Attendance (%)"] >= 75) & (df["Average"] >= 45)]

# Sort by Average descending and pick Top 3
top3 = eligible_df.sort_values(by="Average", ascending=False).head(3)

# Plot
plt.figure(figsize=(8,5))
plt.bar(top3["Name"], top3["Average"], color='green')
plt.title("Top 3 Students by Average Marks")
plt.xlabel("Student Name")
plt.ylabel("Average Marks")
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
